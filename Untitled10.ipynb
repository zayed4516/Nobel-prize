```python
df['usa_born_winner'] = df['birth_country']=='United States of America'
df['decade']=(np.floor(df['year']/10)*10).astype(int)

prop_usa_winners=df.groupby('decade',as_index=False)['usa_born_winner'].mean()

prop_usa_winners
```

# Calculating the proportion of USA born winners per decade

```python
ax = sns.relplot(x='decade', y='usa_born_winner',data=prop_usa_winners,kind='line')
ax
```

# Calculating the proportion of female laureates per decade

```python
df['female_winners']=df['sex']=='Female'
df['decade']=(np.floor(df['year']/10)*10).astype(int)
prop_female_winners=df.groupby(['decade','category'],as_index=False)['female_winners'].mean()
```

```python
prop_female_winners
```

```python
ax = sns.relplot(x='decade', y='female_winners',data=prop_female_winners,kind='line')
ax
```
